```python
import transformers
import torch

# usamos este llamada 3 desde este repo por que si usamos el oficial tenermos que  reguistranos y solicitar el token
model_id = "unsloth/llama-3-8b-Instruct"

# Aca cargamos el modelo en tamano reducido
pipeline = transformers.pipeline(
    "text-generation",  # Tarea de generación de texto
    model=model_id,     # ID del modelo que deseas cargar
    model_kwargs={"torch_dtype": torch.float16, "low_cpu_mem_usage": True},  # Eliminamos quantization
)
```
